````python
# ============================================================================
# KAGGLE CELL: EXPAND GOLDEN for SFT  (generate -> verify -> filter)
#   Input : 1,354 labeled questions (already have answers)
#   Method: generate code WITHOUT showing the answer (no hardcoding possible),
#           execute, verify against the known answer (unit-aware), keep passes.
#   Guards: (1) hold out eval+test (no leakage)  (2) anti-hardcode filter
#           (3) verify vs ground truth           (4) N=3 attempts per question
#   Output: verified_golden_expanded.csv  + holdout_test.csv
# ============================================================================
import os, re, math, signal, random
import pandas as pd
import chromadb
from chromadb.utils import embedding_functions

random.seed(42)
N_ATTEMPTS = 2          # max attempts/question; early-stop on BTC match (usually 1 call)
GEN_TEMPS  = [0.1, 0.6, 0.8]   # temperature per attempt

# ----------------------------------------------------------------------------
# 0. PATHS + CHROMA
# ----------------------------------------------------------------------------
chroma_client = chromadb.PersistentClient(path="/kaggle/working/rag_database")
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2")
collection = chroma_client.get_collection(
    name="physics_golden_knowledge", embedding_function=sentence_transformer_ef)

TEST_FILE   = "/kaggle/input/datasets/quanghuy225/physics/Physics_Problems_Text_Only.csv"
GOLDEN_FILE = "/kaggle/input/datasets/quanghuy225/finalphysic/golden_rag_data_physic_v2.csv"
OUT_EXPANDED = "/kaggle/working/verified_golden_expanded.csv"   # final = golden + new clean
OUT_EXPANDED_NEW = "/kaggle/working/new_clean.csv"             # only newly generated clean (append, resumable)
OUT_TEST     = "/kaggle/working/holdout_test.csv"
OUT_SUSPECT  = "/kaggle/working/suspect_btc.csv"
OUT_FAILED   = "/kaggle/working/generation_failed.csv"

# ----------------------------------------------------------------------------
# 1. LLM
# ----------------------------------------------------------------------------
import ollama
def call_llm(prompt, temperature=0.1):
    resp = ollama.chat(model="qwen2-math:7b",
                       messages=[{"role":"user","content":prompt}],
                       options={"temperature":temperature})
    return resp["message"]["content"]

# ----------------------------------------------------------------------------
# 2. CHECKLIST (reuse v3 - condensed) + topic keyword router (fast, no LLM call)
# ----------------------------------------------------------------------------
UNIVERSAL_CHECKLIST = """
UNIVERSAL CHECKS:
[U1] Convert all quantities to SI before computing (cm->m, uF->F, uC->C, ...).
[U2] Constants: k=9e9, g=10, eps0=8.85e-12.
[U3] Vectors: fix every direction before summing. Field of + charge points AWAY, - charge TOWARD.
     Between OPPOSITE charges fields are CO-DIRECTIONAL -> ADD; between LIKE charges -> SUBTRACT.
[U4] sqrt() argument must be >= 0; denominators != 0.
[U5] Point charge: E = k*q/r**2 (distance SQUARED). Coulomb: F = k*q1*q2/r**2.
[U6] If a magnitude is asked, use abs(). Answer must be a NUMBER (or 'Yes'/'No').
[U7] Variable names: no spaces, no Greek letters.
[U8] Code MUST end with `final_result = <expression>` computed from inputs. NEVER write a literal number.
[U9] Capacitor: source DISCONNECTED -> Q constant; CONNECTED -> U constant.
""".strip()

TOPIC_KEYWORDS=[('ac_resonance',['resonance','rlc','resonant']),
  ('measurement_error',['absolute error','relative error','percent error','uncertainty','measured value']),
  ('LC_oscillation',['lc circuit','oscillation','natural period','electromagnetic oscillation']),
  ('induction',['induced','inductance','self-induction','emf','solenoid','magnetic flux']),
  ('capacitor',['capacitor','capacitance','farad','charge stored','dielectric','plates']),
  ('circuit_power',['power','watt','consumes','dissipat']),
  ('circuit_resistance',['resistor','resistance','ohm','series','parallel','ammeter','current through']),
  ('electrostatics_force',['force on','coulomb force','net force','repuls','attract']),
  ('electrostatics_field',['electric field','field strength','field intensity','v/m','point charge','linear charge'])]
def classify_topic(question):
    q=question.lower()
    for t,kws in TOPIC_KEYWORDS:
        if any(kw in q for kw in kws): return t
    return 'general_physics'

# ----------------------------------------------------------------------------
# 3. RAG retrieve (boost yield with similar solved examples)
# ----------------------------------------------------------------------------
def rag_retrieve(query, k=2):
    try:
        res=collection.query(query_texts=[query], n_results=k)
        docs=res.get("documents",[[]])[0]; metas=res.get("metadatas",[[]])[0]
        blocks=[]
        for d,m in zip(docs,metas):
            blocks.append(f"Q: {d}\nCode:\n{(m or {}).get('golden_code','')}")
        return "\n---\n".join(blocks) if blocks else "(none)"
    except Exception:
        return "(none)"

# ----------------------------------------------------------------------------
# 4. GENERATION PROMPT (answer NOT shown -> cannot hardcode)
# ----------------------------------------------------------------------------
def gen_prompt(question, topic, examples):
    return f"""You are a physics expert. Solve the problem with a Chain-of-Thought and Python code.

PROBLEM:
{question}

TOPIC: {topic}

SIMILAR SOLVED EXAMPLES (reference only; adapt, do not copy):
{examples}

{UNIVERSAL_CHECKLIST}

Output format (BOTH sections):
COT:
1. <step>
2. <step>
3. <step>
CODE:
```python
<code that computes the answer and ends with final_result = <expression>>
```
"""

def extract_code(o):
    m=re.search(r'```python\s*(.*?)```',o,re.DOTALL)
    if m: return m.group(1).strip()
    m=re.search(r'CODE:\s*(.+)',o,re.DOTALL)
    return m.group(1).strip() if m else ""
def extract_cot(o, code):
    m=re.search(r'COT:\s*(.*?)(?:CODE:|```)',o,re.DOTALL)
    if m and m.group(1).strip(): return m.group(1).strip()
    comments=[l.strip('# ').strip() for l in (code or '').split('\n') if l.strip().startswith('#') and len(l.strip())>2]
    return '\n'.join(f"Step {i+1}: {c}" for i,c in enumerate(comments)) if comments else ""

# ----------------------------------------------------------------------------
# 5. SANITIZER + SANDBOX + ANTI-HARDCODE
# ----------------------------------------------------------------------------
GREEK={'κ':'k','π':'pi','θ':'theta','ε':'eps','λ':'lam','ρ':'rho','μ':'mu','ω':'omega',
       'α':'alpha','β':'beta','Δ':'d','σ':'sigma','φ':'phi','Ω':'Ohm','τ':'tau','γ':'gamma'}
def sanitize_code(code):
    if not code: return code
    for g,r in GREEK.items(): code=code.replace(g,r)
    for nm in set(re.findall(r'\b([A-Za-z]\w* +\w+) *=(?!=)',code)):
        code=code.replace(nm, nm.replace(' ','_'))
    return code

def is_hardcoded(code):
    """True if final_result is assigned a literal number (the bug we must avoid)."""
    for m in re.finditer(r'final_result\s*=\s*(.+)', str(code)):
        val=m.group(1).strip().rstrip(';').strip()
        if re.match(r'^-?\d+\.?\d*(?:[eE][+-]?\d+)?$', val):
            return True
    return False

class TimeoutException(Exception): pass
def _h(s,f): raise TimeoutException()
def execute_sandbox(code, timeout=5):
    if not code: return None,"empty"
    code=sanitize_code(code)
    env={'math':math,'pi':math.pi,'sqrt':math.sqrt,'k':9e9,'g':10.0,'eps0':8.85e-12,'e':1.6e-19,'c':3e8}
    try:
        signal.signal(signal.SIGALRM,_h); signal.alarm(timeout)
        exec(code,{'__builtins__':__builtins__},env); signal.alarm(0)
        if 'final_result' not in env: return None,"no_result"
        return env['final_result'],None
    except Exception as ex:
        signal.alarm(0); return None,f"{type(ex).__name__}"

# ----------------------------------------------------------------------------
# 6. ANSWER MATCHING (unit-aware)
# ----------------------------------------------------------------------------
def parse_num(s):
    s=str(s).strip()
    for a,b in {'⁰':'0','¹':'1','²':'2','³':'3','⁴':'4','⁵':'5','⁶':'6','⁷':'7','⁸':'8','⁹':'9','⁻':'-'}.items(): s=s.replace(a,b)
    s=s.replace('{','').replace('}','').replace(' ','')
    m=re.match(r'^(-?\d+\.?\d*)[\.\*xX×e]+10\^?(-?\d+)$',s)
    if m: return float(m.group(1))*10**int(m.group(2))
    m=re.match(r'^10\^?(-?\d+)$',s)
    if m: return 10**int(m.group(1))
    try: return float(s)
    except: return None
SI_SCALES={2,3,6,9,12}
def verify(model_val, gt):
    if model_val is None: return False
    mv=parse_num(model_val); gn=parse_num(gt)
    if mv is not None and gn is not None:
        if mv==0 and gn==0: return True
        if math.isclose(mv,gn,rel_tol=2e-2): return True
        if mv!=0 and gn!=0:
            log=math.log10(abs(mv/gn)); n=round(log)
            if abs(log-n)<0.02 and abs(n) in SI_SCALES: return True
        return False
    return str(model_val).strip().lower()==str(gt).strip().lower()
def _fmt(v):
    if isinstance(v,float):
        if abs(v)>=1e4 or (abs(v)<1e-3 and v!=0): return f"{v:.4e}"
        return f"{v:g}"
    return str(v)

# ----------------------------------------------------------------------------
# 7. CONSENSUS + SANITY (two independent referees: BTC answer + N=3 self-consistency)
# ----------------------------------------------------------------------------
import statistics
from collections import Counter

def consensus(values):
    """Return (representative_value, num_agreeing) for the largest cluster."""
    vals=[parse_num(v) for v in values]; vals=[v for v in vals if v is not None]
    strs=[str(v).strip().lower() for v in values if parse_num(v) is None]
    if len(vals)>=len(strs) and vals:
        clusters=[]
        for v in vals:
            for c in clusters:
                if (v==0 and c[0]==0) or (c[0]!=0 and math.isclose(v,c[0],rel_tol=2e-2)):
                    c.append(v); break
            else: clusters.append([v])
        best=max(clusters,key=len)
        return _fmt(statistics.median(best)), len(best)
    if strs:
        top,cnt=Counter(strs).most_common(1)[0]
        return top, cnt
    return None,0

MAGNITUDE_KW=['magnitude','strength','field intensity','force','energy','power','distance','speed','resistance']
def passes_sanity(value, question):
    n=parse_num(value)
    if n is None: return True
    q=question.lower()
    if any(kw in q for kw in MAGNITUDE_KW) and n<0: return False
    if ('probability' in q or 'efficiency' in q) and (n<0 or n>1): return False
    return True

def generate_classified(question, gt_answer, n=N_ATTEMPTS):
    """Run up to N attempts. EARLY-STOP as soon as one attempt matches BTC (CLEAN).
       Otherwise use consensus to flag SUSPECT_BTC vs REJECT."""
    topic=classify_topic(question)
    examples=rag_retrieve(question)
    outputs=[]  # (value, code, cot) for valid, non-hardcoded attempts
    for i in range(n):
        temp=GEN_TEMPS[min(i,len(GEN_TEMPS)-1)]
        out=call_llm(gen_prompt(question,topic,examples), temp)
        code=sanitize_code(extract_code(out))
        if is_hardcoded(code): continue            # guard: anti-hardcode
        result,err=execute_sandbox(code)
        if err: continue
        val=_fmt(result)
        if not passes_sanity(val, question): continue   # guard: physical sanity
        # EARLY-STOP: matches the independent BTC answer -> CLEAN immediately
        if verify(val, gt_answer):
            return {"tier":"CLEAN","reason":"match_btc","code":code,
                    "cot":extract_cot(out,code),"computed":val,"agree":1,
                    "topic":topic,"attempts":i+1}
        outputs.append((val, code, extract_cot(out,code)))

    if not outputs:
        return {"tier":"REJECT","reason":"no_valid_code","topic":topic,"attempts":n}
    # no attempt matched BTC -> is the model at least self-consistent?
    values=[o[0] for o in outputs]
    cons_val, agree = consensus(values)
    if agree >= 2:    # model agrees on a value != BTC -> BTC likely wrong
        rep=next((o for o in outputs if o[0]==cons_val or
                  (parse_num(o[0]) is not None and parse_num(cons_val) is not None
                   and parse_num(cons_val)!=0 and math.isclose(parse_num(o[0]),parse_num(cons_val),rel_tol=2e-2))),
                 outputs[0])
        return {"tier":"SUSPECT_BTC","reason":"consensus_vs_btc_mismatch","code":rep[1],
                "cot":rep[2],"computed":cons_val,"agree":agree,"topic":topic,"attempts":n}
    return {"tier":"REJECT","reason":"no_consensus","topic":topic,"computed":cons_val,"attempts":n}

# ----------------------------------------------------------------------------
# 8. SPLIT (hold out eval 35 + test 150; no leakage)
# ----------------------------------------------------------------------------
test=pd.read_csv(TEST_FILE); golden=pd.read_csv(GOLDEN_FILE)
test['prefix']=test['id'].astype(str).str.extract(r'^([A-Za-z]+)')
golden_ids=set(golden['id'].astype(str))
test=test[test['prefix']!='QA']
unseen=test[~test['id'].astype(str).isin(golden_ids)]
PREFIXES=['CH','NL','LD','DDT','TD','THCB','DT']

eval_ids=set()
for p in PREFIXES:
    pool=unseen[unseen['prefix']==p]
    eval_ids|=set(pool.sample(n=min(5,len(pool)),random_state=42)['id'].astype(str))
remain=unseen[~unseen['id'].astype(str).isin(eval_ids)]
test_ids=set()
for p in PREFIXES:
    pool=remain[remain['prefix']==p]
    n=min(round(150*len(pool)/len(remain)),len(pool))
    test_ids|=set(pool.sample(n=n,random_state=123)['id'].astype(str))
train_pool=remain[~remain['id'].astype(str).isin(test_ids)].reset_index(drop=True)

# Save holdout test
holdout=unseen[unseen['id'].astype(str).isin(eval_ids|test_ids)][['id','prefix','question','answer','unit']]
holdout.to_csv(OUT_TEST,index=False)
print(f"Hold-out test saved: {len(holdout)} | Train pool: {len(train_pool)}")

# ----------------------------------------------------------------------------
# 9. RUN GENERATION  (RESUMABLE + incremental save + time budget)
# ----------------------------------------------------------------------------
import time
MAX_RUNTIME_SEC = 11*3600   # stop & exit gracefully before Kaggle's 12h limit
start_time = time.time()

def append_row(path, row_dict):
    """Append one row to CSV immediately (write header if file is new)."""
    df1 = pd.DataFrame([row_dict])
    df1.to_csv(path, mode='a', header=not os.path.exists(path), index=False)

# RESUME: collect IDs already processed in any output file
done_ids = set()
for p in [OUT_EXPANDED_NEW, OUT_SUSPECT, OUT_FAILED]:
    if os.path.exists(p):
        try: done_ids |= set(pd.read_csv(p)['id'].astype(str))
        except Exception: pass
print(f"Resume: {len(done_ids)} questions already done, skipping them.")

processed = 0
for idx, row in train_pool.iterrows():
    if str(row['id']) in done_ids:        # RESUME: skip finished
        continue
    if time.time() - start_time > MAX_RUNTIME_SEC:
        print(f"\n[TIME BUDGET] Stopping at {processed} new questions. Re-run to continue.")
        break
    res = generate_classified(row['question'], row['answer'])
    base = {"id":row['id'],"prefix":row['prefix'],"question":row['question'],
            "answer":row['answer'],"unit":row.get('unit','')}
    if res["tier"]=="CLEAN":
        append_row(OUT_EXPANDED_NEW, {**base,"golden_code":res["code"],"cot":res["cot"],
                                      "topic":res["topic"],"agree":res["agree"]})
    elif res["tier"]=="SUSPECT_BTC":
        append_row(OUT_SUSPECT, {**base,"model_consensus":res["computed"],"agree":res["agree"],
                                 "golden_code":res["code"],"cot":res["cot"],"topic":res["topic"]})
    else:
        append_row(OUT_FAILED, {**base,"reason":res["reason"]})
    processed += 1
    if processed % 10 == 0:
        el=(time.time()-start_time)/60
        print(f"  +{processed} new ({el:.0f} min) | last={res['tier']} ({res.get('attempts','?')} calls)")

# ----------------------------------------------------------------------------
# 10. COMBINE clean (new + existing golden) -> final SFT file
# ----------------------------------------------------------------------------
g=golden.copy()
if 'cot' not in g.columns: g['cot']=''
if 'topic' not in g.columns: g['topic']=g['question'].apply(classify_topic)
keep=['id','prefix','question','golden_code','cot','answer','unit','topic']
for c in keep:
    if c not in g.columns: g[c]=''
parts=[g[keep]]
if os.path.exists(OUT_EXPANDED_NEW):
    nd=pd.read_csv(OUT_EXPANDED_NEW)
    for c in keep:
        if c not in nd.columns: nd[c]=''
    parts.append(nd[keep])
combined=pd.concat(parts, ignore_index=True).drop_duplicates(subset='id')
combined.to_csv(OUT_EXPANDED, index=False)

# ----------------------------------------------------------------------------
# 11. SUMMARY
# ----------------------------------------------------------------------------
def _count(p): return len(pd.read_csv(p)) if os.path.exists(p) else 0
nc, ns, nr = _count(OUT_EXPANDED_NEW), _count(OUT_SUSPECT), _count(OUT_FAILED)
print("\n"+"="*60)
print(f"This run processed   : {processed} new questions")
print(f"  CLEAN (cumulative) : {nc}")
print(f"  SUSPECT_BTC        : {ns}  <- review, report to BTC (Q22 bonus)")
print(f"  REJECT             : {nr}")
print(f"Existing clean golden: {len(golden)}")
print(f"TOTAL expanded golden: {len(combined)}  -> {OUT_EXPANDED}")
print(f"\nIf not finished: just RE-RUN this cell - it resumes (skips done IDs).")
print("Saved incrementally, so a timeout no longer loses progress.")
````
